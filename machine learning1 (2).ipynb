```python
df_encoded=pd.get_dummies(df,columns=['Gender','Country','Most_Used_Platform','Relationship_Status'],dtype=int)
print(df_encoded)
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df["Gender"]=le.fit_transform(df["Gender"])
df["Country"]=le.fit_transform(df["Country"])
df["Most_Used_Platform"]=le.fit_transform(df["Most_Used_Platform"])
df["Affects_Academic_Performance"]=le.fit_transform(df["Affects_Academic_Performance"])
df["Relationship_Status"]=le.fit_transform(df["Relationship_Status"])
df["Academic_Level"]=le.fit_transform(df["Academic_Level"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
for col in df:
    plt.figure(figsize=(4,6))
    sns.distplot(df[col])
    plt.show()
```

```python
#split the data in independent and independent variables
X=df[['Avg_Daily_Usage_Hours','Affects_Academic_Performance','Sleep_Hours_Per_Night','Mental_Health_Score','Conflicts_Over_Social_Media']]
Y=df['Addicted_Score']
```

```python
X
```

```python
#spliting the data in train and test

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=23)
```

```python
#apply model
#regression model
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

model1=lr.fit(X_train,Y_train)

y_prediction1=lr.predict(X_test)
```

```python
y_prediction1
```

```python
from sklearn.metrics import r2_score
```

```python
score1=r2_score(Y_test,y_prediction1)
score1
```

```python
from sklearn.linear_model import Ridge
rd=Ridge()
model2=rd.fit(X_train,Y_train)
y_prediction2=rd.predict(X_test)
```

```python
y_prediction2
```

```python

score2=r2_score(Y_test,y_prediction2)
score2
```

```python
from sklearn.linear_model import Lasso
rd1=Lasso()
model3=rd1.fit(X_train,Y_train)
y_prediction3=rd1.predict(X_test)
```

```python
score3=r2_score(Y_test,y_prediction3)
score3
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
model4=dt.fit(X_train,Y_train)
y_prediction4=dt.predict(X_test)
```
